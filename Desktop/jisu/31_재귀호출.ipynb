```python
# 재귀호출 사용하기
def hello(count):
    if count == 0:
        return
    
    print("hello, world", count)

    count -= 1
    hello(count)
```

```python
hello(5)
```

```python
# 재귀호출로 팩토리얼 구하기

def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)
```

```python
print(factorial(5))
```

```python
# 연습 @@@@@이해못함
def is_palindrome(word):
    if len(word) < 2: # 글자가 없거나 한글자면 같다고 볼 수 있다
        return True
    if word[0] != word[-1]:
        return False
    return is_palindrome(word[1:-1])

print(is_palindrome('hello'))
print(is_palindrome('level'))
```
