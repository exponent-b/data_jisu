```python
# 튜플 사용하기
person = (
    "james", 17, 175.3, True
)
print(person)
```

```python
# 요소가 한 개 들어가 있는 튜플 만들기
a = 38
b = (38,)
c = 38,
print(a)
print(type(a))
print(b)
print(type(b))
print(c)
print(type(c))
```

```python
#튜플을 리스트로 만들고 리스트를 튜플로 만들기
a = [1, 2, 3]
print(a)
print(type(a))
a = tuple(a)
print(a)
print(type(a))
```
