```python
# else와 들여쓰기
x = 5
if x == 10:
    print("10입니다")
else:
    print("x에 들어있는 숫자는")
    print("10이 아닙니다.")
```

```python
# if숫자 지정하기

if 0:
    print("참")

else:
    print("거짓")

if 1:
    print("참")

else:
    print("거짓")

if 0x1f:
    print("참")

else:
    print("거짓")

if 0b1000:
    print("참")

else:
    print("거짓")

if 13.5:
    print("참")

else:
    print("거짓")
```

```python
# 조건문의 문자열 지정하기
if "hello":
    print("참")

else:
    print("거짓")

if ' ':
    print("참")

else:
    print("거짓")

if '':
    print("참")

else:
    print("거짓")

```

```python
# 조건식을 여러개 지정하기
x = 10
y = 20

if (x == 10) and (y == 20):
    print("참")

else:
    print("거짓")
```
