```python
# 심사
text = """the grown-ups' response, this time was to advise me to lay aside 
whether from the inside or the outside and devote myself instead to geography, history, arithmetic, 
my drawings of boa constrictors, and grammar. That is why, at the, age of six. T gave un what might have been a magnificent career as Grown-ups never understand anything by themselves, and it is tiresome for children to be always and a painter. 
I had been disheartened by the failure of my Drawing Number One and my Drawing Number TwO. 
forever explaining things to the."""

text_list = text.split(" ")
#print(text_list)
count = 0
for ch in text_list:
    re = ch.strip(".,-=!*@#?/")
    if re == "the":
        #print(ch)
        count += 1
    else:
        pass

print(count)
```

```python
#answer = list(map(int, input().split(";")))
answer = "51900;83000;158000;367500;250000;59200;128500;1304000"
answer = list(map(int, answer.split(";")))

for i in range(len(answer)):
    mm = max(answer)
    print('{0:>9,}'.format(mm)) #### 
    #print(answer)
    answer.remove(max(answer))
```
