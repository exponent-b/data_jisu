```python
# for 반복문과 if 조건문을 여러번 사용하기
gu = [j * i for i in range(2, 10) for j in range(1, 10)]
print(gu)

gu = []

for i in range(2, 10):
    for j in range(1, 10):
        gu.append(i * j)

print(gu)
```

```python
###
def min(a):
    small = a[0]

    for i in a:
        if i < small:
            small = i
        else:
            pass
    return small

a = [38, 21, 53, 62, 19]
re = min(a)
print(re)
```
