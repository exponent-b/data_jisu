```python
# 파일에 문자열 쓰기, 읽기
# 파일객체 = open(파일이름, 파일모드)
# 파일객체.write("문자열")
# 파일객체.close()

file = open("./data/hello.txt", "w")
file.write("hello world!")
file.close()
```

```python
## 파일에서 문자열 읽기
f = open("./data/hello.txt", "r")
s = f.read()
print(s)
f.close()
```

```python
# 자동으로 파일 객체 닫기 
# with open(파일이름, 파일모드) as 파일객체:
#     코드
with open("data/hello.txt", "r") as file:
    s = file.read()
    print(s)
```

```python
# 문자열 여러줄 파일에 쓰기, 읽기
with open("./data/hello.txt", "w") as file:
    for i in range(3):
        file.write(f"hello world {i} \n")
```

```python
lines = [
    "안녕하세요. \n",
    "파이썬 \n",
    "코딩 도장입니다. \n"    
]
with open("./data/hello.txt", "w") as file:
    file.writelines(lines)
```

```python
## 파일의 내용을 한줄씩 리스트로 가져오기
with open("./data/hello.txt", "r") as file:
    lines = file.readlines()
    print(lines)
```

```python
## 파일의 내용을 한줄씩 읽기
with open("./data/hello.txt", "r") as f:
    line = None
    
    while line != "":
        line = f.readline() # readlines 내에 \n이 있다.
        print(line)
        # print(line.strip())
```

```python
## for반복문으로 파일의 내용을 줄 단위로 읽기
with open("./data/hello.txt", "r") as f :
    for line in f:
        print(line.strip('\n'))
```

```python
## 파일에서 파이썬 객체를 저장하기, 가져오기
import pickle

name = "james"
age = 17
address = "서울시 서초구 반포동"
scores = {
    "kor" : 90,
    "eng" : 95,
    "math" : 85,
    "sci" : 82
}

with open("./data/james.p", "wb") as f:
    pickle.dump(name, f)
    pickle.dump(age, f)
    pickle.dump(address, f)
    pickle.dump(scores, f)
```

```python
import pickle

with open("./data/james.p", "rb") as f:
    name = pickle.load(f)
    age = pickle.load(f)
    address= pickle.load(f)
    scores = pickle.load(f)

    print(name)
    print(age)
    print(address)
    print(scores)
```

```python
# 연습
with open("./data/words.txt", "r") as f:
    count = 0
    for line in f:
        if len(line) <= 10:
            count += 1
    

print(count)
```

```python
# 실전
file = open("./data/words2.txt", "r")
s = file.read()
words = s.split()
for word in words:
    word = word.strip(".,")
    for w in word:
        if w == "c":
            print(word) 

```

```python
# 파일을 읽는 법
print("=====with=====")
with open("./data/hello.txt") as file:
    for f in file:
        print(f, end = "")

print("====read====")
file = open("./data/hello.txt", "r")
s = file.read()
print(s)
file.close()

print("=====readlines=====")
file = open("./data/hello.txt", "r")
s = file.readline() # 한줄만 읽기 때문에 다 읽으려면 반복문 써야함
print(s)
```
