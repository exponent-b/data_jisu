```python
for i in range(5):
    for j in range(5):
        print("j", j, sep = '-', end = ' ')
    print("i:", i, "\\n", sep = '')
```

```python
# 사각형으로 별 출력하기
for i in range(5):
    for j in range(5):
        print("*", end = '')
    print()
```

```python
for i in range(3):
    for j in range(7):
        print("*", end = "")
    print()
```

```python
# 계단식으로 별 출력하기 *********************시험
for i in range(5):
    for j in range(5):
        if j <= i:
            print("*", end = '')
    print()
```

```python
# 대각선으로 별 출력하기
for i in range(5):
    for j in range(5):
        if j == i :
            print("*", end = '')
        else:
            print(" ", end = "")
    print()
```

```python
# 심사
num = int(input())

for i in range(num): # 줄
    for j in range(num - 1 - i): # 공백
        print(" ", end = "")
    print("*" * (i * 2 + 1))        
```

```python
num = 5

for i in range(1, num + 1):
    space = num - i
    star = 2 * i - 1
    print(" " * space + "*" * star)
```
