```python
# 위치 인수와 리스트 언패킹
## 위치 인수를 사용하는 함수
def print_numbers(a, b, c):
    print(a)
    print(b)
    print(c)
```

```python
print_numbers(10, 20, 30) # 변수의 개수가 맞지 않으면 오류가 난다
```

```python
x = [10, 20, 30]
print(x)
print(*x) # 포인터 비슷..?
# *[10, 20, 30]
print(10, 20, 30)
```

```python
## 가변 인수 함수 만들기

def print_numbers(*args):
    for arg in args:
        print(arg)
```

```python
print_numbers("1:"10)

print_numbers(10, 20, 30, 40) # args를 쓰면 개수 상관없이 넣을 수 있다
```

```python
x = [10]
y = [10, 20, 30, 40]

print_numbers("xx: ", *x)
print_numbers("yy:", *y)
```

```python
# 키워드 인수 사용하기
def personal_info(name, age, address):
    print("이름:", name)
    print("나이:", age)
    print("주소:", address)
```

```python
personal_info('홍길동', 30, '서울시 용산구')
```

```python
personal_info(name= '홍길동', address= '서울시 용산구', age = 30)
```

```python
personal_info('홍길동', address= '서울시 용산구', age = 30)
```

```python
# 키워드 인수와 딕셔너리 언패킹 사용하기
x = {
    'name': '홍길동',
    'age': 30,
    'address':'서울시 용산구'
}

personal_info(**x)
```

```python
# 키워드 인수를 사용하기 가변인수 함수 만들기
## * 별이 하나 == 리스트, ** 별이 둘 == 딕셔너리

def personal_info(**kwargs):
    # print(kwargs) #=>{'name': '홍길동', 'age': 30, 'address': '서울시 용산구'}
    # print(type(kwargs)) #=> <class 'dict'>
    for kw, arg in kwargs.items():
        print(f"{kw} : {arg}")
```

```python
personal_info(name ='홍길동', age = 30, address = '서울시 용산구')
```

```python
x = {
    'name': '홍길동'
}

y = {
    'name': '홍길동',
    'age': 30,
    'address':'서울시 용산구'
}
```

```python
personal_info(**x)
personal_info(**y)
```

```python
# 연습
 
korean, english, mathematics, science = 100, 86, 81, 91

def get_max_score(*args):
    m = max(args)
    return m


max_score = get_max_score(korean, english, mathematics, science)
print('높은 점수:', max_score)

max_score = get_max_score(english, science)
print('높은 점수:', max_score)
```

```python
# 실전
korean, english, mathematics, science = map(int, input().split())

def get_min_max_score(*args):
    max_s = max(args)
    min_s = min(args)

    return min_s, max_s


def get_average(**args):
    hap = 0
    for a in args.values(): # 딕셔너리형이므로 값을 부르는 value()를 써야한다
        hap += (a)
    result = hap / len(args)
    return result



min_score, max_score = get_min_max_score(korean, english, mathematics, science)
average_score = get_average(korean=korean, english=english, mathematics=mathematics, science=science) ## 딕셔너리로 써야함
print('낮은 점수: {0:.2f}, 높은 점수: {1:.2f}, 평균 점수: {2:.2f}'.format(min_score, max_score, average_score))

min_score, max_score = get_min_max_score(english, science)
average_score = get_average( english=english, science=science)
print('낮은 점수: {0:.2f}, 높은 점수: {1:.2f}, 평균 점수: {2:.2f}'.format(min_score, max_score, average_score))
```
