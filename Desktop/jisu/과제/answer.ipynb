```python
##1

students = {
    "김철수": [90, 80, 70],
    "이영희": [100, 95, 90],
    "박민수": [60, 70, 65]
}

hap_avg = 0
for student in students.keys():
    print("이름:", student)
    score = students[student]
    print("점수:", score)
    s_avg = sum(score) / len(score)
    print("평균:", s_avg)
    hap_avg += s_avg
    print()

print("===== 전체 결과 =====")
print(f"전체 학생 수: {len(students)}명")
print("전체 평균:", hap_avg / len(students))
```

```python
## 6
'''
두 수 사이의 숫자 중  3과 7과 끝나지 않는 숫자 출력하기 

입력값 
3, 50

첫 번째 정수와 두 번째 정수 사이의 숫자 중 
3 또는 7로 끝나지 않늣 숫자가 출력되게 만드세요 
'''

num1, num2 = map(int, input().split(", "))

for i in range(num1, num2 + 1):
    if i % 10 == 3 or i % 10 == 7:
        pass
    else:
        print(i,  end = ' ')
```

```python
## 7
'''
표준 입력으로 역삼각형의 높이가 입력됩니다. 
입력된 높이만큼 역삼각형의 "*"을 출력하는 프로그램을 작성하세요 

입력갑
5

출력값 
*********
 *******
  *****
   ***
    *

요구사항:
중첩 for를 사용
'''
num1 = int(input())

for i in range(num1):
    print(" " * i, end = '')
    print("*" * ((num1 - i) * 2 - 1))
```

```python
## 8
'''
입력값을 참조하여 첫 번째 정수 부터 두 번째 정수까지 
숫자를 역으로 출력하면서 
3의 배수일때는 Fizz, 7의 배수일때는 Buzz, 그외는 숫자를 출력하세요 

입력값 
100 10
'''
num1, num2 = map(int, input().split())

for i in range(num1, num2 - 1, -1):
    if i % 3 == 0 and i % 7 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 7 == 0:
        print("Buzz")
    else:
        print(i)
```

```python
## 9
'''
표준 입력으로 2개의 정수가 입력된다. 
첫번째 수와 두 번째 수 사이에 제곱근을 
리스트(소수점 첫째자리)로 출력하는 프로그램을 만드세요 

입력값
1000 1200

요구사항
1. 리스트 컴프리헨션 표현식과 일반표현식으로 작성
(2가지 방법)
'''
# 일반표현식

num1, num2 = map(int, input().split())
a_list = []

for i in range(num1, num2 + 1):
    result = float(format(i ** 0.5, ".1f"))
    a_list.append(result)

print(a_list)

```

```python
# 리스트 컴프리헨션 표현식
num1, num2 = map(int, input().split())
a_list = [float(format(i ** 0.5, ".1f")) for i in range(num1, num2 + 1)]
print(a_list)
```

```python
## 10
'''
표준 입력으로 2차원 리스트의 가로(col)와 세로(row)가 입력되고 
그 다음 줄부터 리스트의 요소로 들아갈 문자가 입력됩니다. 
이때 2차원 리스트 안세어 *는 지뢰이고, .은 지뢰가 압니다. 
지뢰가 아닌 요소에는 인접한 지뢰의 개수를 출력하는 프로그램을 만드세요 

matrix = [] 
for i in range(row): 
    matrix.append(list(input())) 


입력 
3 3
.**
*..
.*.

출력
2**
*43
2*1
'''
```

```python
col, row = map(int, input().split())

matrix = [] 
for i in range(row): 
    matrix.append(list(input())) 

print(matrix)
```

```python
for r in range(row):
    for c in range(col):

        if matrix[r][c] == "*":
            continue
        
        else:
            count = 0
            for rr in range(r -1, r + 2):
                for cc in range(c - 1, c + 2):
                    if 0 <= cc < col and 0 <= rr < col:
                        if matrix[rr][cc] == "*":
                            count += 1
            matrix[r][c] = count

#print(matrix)

for i in range(row):
    for j in range(col):
        print(matrix[i][j], end = "")
    print()
        
```

```python
## 11
'''
표준 입력으로 물품 가격 여러 개가 문자열 한 줄로 입력되고, 
각 가격은(:) 콜론으로 구분되어 있음
입력된 가격은 높은 가격순으로 출력하는 프로그램을 만드세요 

입력
51900:12344:79900:13433:88888:1352423:9023313
'''
price = list(map(int, input().split(":")))
for i in range(len(price)):
    print(max(price))
    price.remove(max(price))
```

```python
## 12
'''
표준 입력으로 첫 번째 줄은 키, 두 번째 줄은 값으로 하여 딕셔너리르 생성하고 
딕셔너리에 키가 delta인 키-값 쌍과 값이 30인 키-값 쌍을 삭제하도록 만드세요 

alpha bravo charlie delta
10 20 30 40
'''
key_num = list(input().split())
val_num = list(input().split())
dic = dict(zip(
        key_num, val_num
    ))

dic.pop("delta")

for key in list(dic):
    if dic[key] == "30":
        del dic[key]

print(dic)
```
