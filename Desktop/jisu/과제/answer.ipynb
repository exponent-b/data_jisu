```python
students = {
    "김철수": [90, 80, 70],
    "이영희": [100, 95, 90],
    "박민수": [60, 70, 65]
}

hap_avg = 0
for student in students.keys():
    print("이름:", student)
    score = students[student]
    print("점수:", score)
    s_avg = sum(score) / len(score)
    print("평균:", s_avg)
    hap_avg += s_avg
    print()

print("===== 전체 결과 =====")
print(f"전체 학생 수: {len(students)}명")
print("전체 평균:", hap_avg / len(students))
```
