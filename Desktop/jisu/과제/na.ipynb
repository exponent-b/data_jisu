```python
'''
문제 10. 
표준 입력으로 2차원 리스트의 가로(col)와 세로(row)가 입력되고 그 다음 줄부터 리스트의 요소로 들아갈 문자가 입력됩니다. 
이때 2차원 리스트 안세어 *는 지뢰이고, .은 지뢰가 압니다. 지뢰가 아닌 요소에는 인접한 지뢰의 개수를 출력하는 프로그램을 만드세요 

matrix = [] 
for i in range(row): 
    matrix.append(list(input())) 


입력 
3 3
.**
*..
.*.

출력
2**
*43
2*1
'''
col, row = 3, 3
matrix = []
for i in range(row): 
    matrix.append(list(input())) 
```

```python
for r in range(row):
    for c in range(col):
        if matrix[r][c] == "*":
            pass

        else:
            count = 0
            for rr in range(r - 1, r + 2):
                for cc in range(c - 1, c + 2):
                    if 0 <= rr < row and 0 <= cc < col: # <= row아님!!!!!111
                        if matrix[rr][cc] == '*':
                            count += 1
                        else:
                            pass
            matrix[r][c] = count 
#print(matrix)    

#or
#for i in range(row):
#    for j in range(col):
#        print(matrix[i][j], end = '')
#    print()

for i in matrix:
    for j in i:
        print(j, end = "")
    print()
            
```

```python
''' @@@@@
문제 11. 
표준 입력으로 물품 가격 여러 개가 문자열 한 줄로 입력되고, 각 가격은(:) 콜론으로 구분되어 있음
입력된 가격은 높은 가격순으로 출력하는 프로그램을 만드세요 

입력
51900:12344:79900:13433:88888:1352423:9023313
'''

line = "51900:12344:79900:13433:88888:1352423:9023313"

line = list(map(int, line.split(":")))

for i in range(len(line)):
    ma = max(line)
    print(ma)
    line.remove(ma)

# pop안된다 제발
```

```python
''' @@@@@@@@@@@@@2
문제 12.
표준 입력으로 첫 번째 줄은 키, 두 번째 줄은 값으로 하여 딕셔너리르 생성하고 
딕셔너리에 키가 delta인 키-값 쌍과 값이 30인 키-값 쌍을 삭제하도록 만드세요 

alpha bravo charlie delta
10 20 30 40
'''

n1 = "alpha bravo charlie delta"
n2 = "10 20 30 40"

n1 = list(n1.split())
n2 = list(map(int, n2.split()))

dic = dict(zip(
    n1, n2
))

#dic.pop("delta")
del dic["delta"]

for key, val in dic.copy().items():
    if val == 30:
        #del dic[key]
        dic.pop(key)

print(dic)
```
