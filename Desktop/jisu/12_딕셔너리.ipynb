```python
# 딕셔너리의 키에 접근하고 값 할당하기
lux = {'h': 490, 'a': 450, 'ddd': 887}
lux['h'] = 2037
lux['a'] = 1104
lux['mmr'] = 2000
print(lux)
```

```python
lux.get("att", 0)
```

```python
# 딕셔너리에 키가 있는지 확인하기
print('h' in lux)
print(2000 in lux)
```

```python
# 딕셔너리 키의 개수 구하기
print(len(lux))
```
