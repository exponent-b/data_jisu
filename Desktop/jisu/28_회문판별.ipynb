```python
# 회문판별하기
## 반복문으로 문자 검사하기

word = input("단어를 입력하세요: ")

is_palindrome = True

for i in range(len(word)):
    if word[i] != word[-1 -i]:
        is_palindrome = False
        break
print(is_palindrome)
```

```python
## 시퀀스 뒤집기로 문자 검사하기
word = input("단어를 입력하세요: ")
print(word == word[::-1]) 
```

```python
## 리스트와 reverse 사용하기
word = "level"

list(word) == list(reversed(word))
```

```python
## 문자열의 join매서드와 reversed 사용하기

word = 'level'

print(word == "".join(reversed(word)))
# print(word == reversed(word))
```

```python
# N-gram 만들기
## 반복문으로 N-gram 출력하기

text = "hello"

for i in range(len(text) - 1):
    print(text[i], text[i + 1], sep = '')
```

```python
text = "this is python script"
words = text.split()
words
```

```python
## zip으로 2-gram만들기
text = "hello"

two_gram = zip(text, text[1:])
# list(two_gram)
# [('h', 'e'), ('e', 'l'), ('l', 'l'), ('l', 'o')]

for i in two_gram:
    print(i[0], i[1], sep = '')

# he
# el
# ll
# lo
```

```python
## zip과 리스트 표현식으로 N-gram만들기
text = "hello"

[ text[i:] for i in range(3) ]
```

```python
# list(zip(*['hello', 'ello', 'llo'])) *가 아래와 같이 바뀜
list(zip("hello", "ello", "llo"))
```
