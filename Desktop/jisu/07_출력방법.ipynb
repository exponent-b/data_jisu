```python
# 값을 여러 개 출력하기
print(1, 2, 3,)
print("Hello", "python")
```

```python
# sep로 값 사이에 문자 넣기
print(1, 2, 3, sep = ', ')
print(4, 5, 6, sep = '')
print(1920, 1000, sep = "x")
```

```python
# 줄바꿈 활용하기
print(1, 2, 3, sep = "\n")
print("1\n2\n3\n")
```

```python
# and 사용하기
print(1, end ='')
print(2, end ='')
print(3, end ='')
```

```python
# sep 값 사이에 넣기
# end 줄자꿈 대신 넣기
print(1, 2, 3, sep = "--", end = "/")
print(4, 5, 6, sep = "**")
```

```python
year, month, day, hour, min, sec = input().split()
print(year, month, day, sep = "-", end = " T ->")
print(hour, min, sec, sep = ":")
```
