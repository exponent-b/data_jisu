```python
# while반복문을 한 번에 사용하기
a = [
    [10, 20],
    [30, 40],
    [50, 60]
]
i = 0
while i < len(a):
    x, y = a[i]
    print(x, y)
    i += 1
```

```python
# while반복문을 두 번 사용하기
a = [
    [10, 20],
    [30, 40],
    [50, 60]
]

i = 0
while i < len(a):
    j = 0
    while j < len(a[i]):
        print(a[i][j], end = ' ')
        j += 1
    print()
    i += 1
```

```python
# 심사
col, row = map(int, input().split())
mat = []
safe = [[0 for _ in range(col)] for _ in range(row)]

#print(safe)

for i in range(row): # 지뢰 세로 입력
    mat.append(list(input()))

# 지뢰* 안전.
for i in range(len(mat)):
    for j in range(len(mat[i])):
        check = mat[i][j]

        #줄에 지뢰 있니
        if check == '*':
            safe[i][j] = "*"
            pass
        #없으면 계산
        elif check == '.':
            for a in range(i-1, i+2):
                for b in range(j-1, j+2): # 넘어가면 오류...
                    if 0 <= a <= row -1 and 0 <= b <= col -1: # 
                        if mat[a][b] == ".":
                            #print(a, b)
                            pass

                        elif mat[a][b] == "*":
                            safe[i][j] += 1

                        #print("1", safe)

#print(safe)
for one in range(row):
    for two in range(col):
        print(safe[one][two], end = ' ')
    print()
```

```python
row, col = map(int, input().split())
matrix = []

for i in range(row): # 지뢰 입력
    matrix.append(list(input()))

print(matrix)
```

```python
for r in range(row):
    for c in range(col):

        if matrix[r][c] == "*":
            continue

        count = 0
        for nr in range(r-1, r+2):
            for nc in range(c-1, c+2):
                if 0 <= nr < row and 0 <= nc < col:
                    if matrix[nr][nc] == "*":
                        count += 1
    
        matrix[r][c] = str(count)

matrix
```
