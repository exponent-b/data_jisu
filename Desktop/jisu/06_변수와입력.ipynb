```python
# 입력값을 정수로 반환하기
a = int(input("첫 번째 숫자를 입력하세요: "))
b = int(input("두 번째 숫자를 입력하세요: "))
print(a)
print(b)

print(a + b)
```

```python
# 입력값을 변수 두 개에 저장하기
a, b = input("문자열 두 개를 입력하세요: ").split()
print(a)
print(b)
```

```python
a, b = map(int, input("숫자 두 개를 입력하세요.").split())
print(a)
print(b)
print(a + b)
```

```python
# 입력받은 값을 콤마를 기분으로 분리하기
a, b = map(int, input("숫자 두 개를 입력하세요.").split(","))
print(a + b)
```
