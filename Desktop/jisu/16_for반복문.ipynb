```python
# for 변수 in range(횟수):
#     반복할 코드

for i in range(10):
    print(f"hello{i}")
```

```python
for _ in range(10):
    print("hello")
```

```python
# 시작하는 숫자와 끝나는 숫자 지정하기
for i in range(5, 12):
    print("hello", i)
```

```python
# 증가폭 사용하기
for i in range(0, 10, 2):
    print("hello", i)
```

```python
# 숫자를 감소시키기
for j in range(10, 0, -1):
    print("hello", j)
```

```python
for i in reversed(range(10)):
    print(i)
```

```python
for i in reversed(range(-5, 10, 2)):
    print(i)

# range는 원래 쓰는 것처럼 (10, -5, 2) 반대는 안나옴
```

```python
# 입력한 횟수대로 반복하기
count = int(input("반복할 횟수: "))
for i in range(count):
    print("hello", i)
```

```python
# 시퀀스 객체로 반복하기
a = [10, 20, 30, 40, 50]
for i in a :
    print(i)
```

```python
## 튜플 
fs = (
    'a', 'b', 'c'
)

for f in fs:
    print(f)
```

```python
for letter in "python":
    print(letter, end = '')

for letter in "python":
    print(letter, end = ' ')

```

```python
print(list(reversed("python")))
```

```python
for letter in reversed("python"):
    print(letter, end = '')
```

```python
x = [49, -17, 25, 102, 8, 62, 21]

for i in range(len(x)):
    x[i] *= 10
    print(x[i], end = ' ')
```

```python
num = int(input("단을 입력하세요: "))
print(f"-----{num}단-----")

for i in range(1, 10):    
    print(f"{num} X {i} = {num * i}")
```
