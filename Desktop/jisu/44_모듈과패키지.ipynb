```python
# import 모듈 가져오기

# import 모듈
# import 모듈1, 모듈2

# 모듈.변수
# 모듈.함수()
# 모듈.클래스()

import math

print(math.pi)
print(math.sqrt(4.0))
```

```python
## as로 모듈이름 지정하기
# import 모듈 as 이름
import math as m

print(m.pi)
print(m.sqrt(4.0))
```

```python
# from import로 모듈의 일부분만 가져오기
from math import pi, sqrt

print(pi)
print(sqrt(2.0))
```

```python
# from 모듈 import * //모든 변수, 함수, 클래스를 들고옴
from math import * # 부담이 되어서 특정 함수만 들고옴
```

```python
# from import로 모듈의 일부를 가져온 뒤 이름 정하기
from math import sqrt as s, pi as p

print(p)
print(s(4.0))
```
