```python
# 함수 안에서 전역 변수 변경하기
b = 10
def foo():
    b = 20
    print(b)

foo()
print(b)
```

```python
c = 10
def foo():
    global c # 함수 안과 밖에서도 쓸 수 있다//재할당이 됐다
    c = 20
    print(c)

foo()
print(c)
```

```python
def foo():
    global d 
    d = 20
    print(d)

foo()
print(d)
```

```python
# 함수 안에서 함수 만들기

# def 함수이름():
#     코드
#     def 함수이름():
#         코드

def print_hello():
    hello = "Hello, World"
    def print_message(): #1실행
        print(hello)  #1반환
    print_message() #1호출

print_hello()
```

```python
# 지역 변수 변경하기
def a(): 
    x = 10 #####
    def b():
        x =20
    b()
    print(x)
a()
```

```python
def a():
    x = 10
    def b():
        nonlocal x
        x = 20
    b()
    print(x)
a()
```

```python
# nonlocal이 변수를 찾는 순서
def a():
    x = 10 # 
    y = 100

    def b():
        x = 20 # 
        def c():
            nonlocal x # b, c의 x와 공유
            nonlocal y
            x += 30
            y += 300
            print(x)
            print(y)

        c()
    b()   
a()
```

```python
# global로 전역 변수 찾기
x = 1
def a():
    x = 10
    def b():
        x = 20
        def c():
            global x
            x += 30
            print(x)
        c()
    b()
a()
```

```python
# 클로저 사용하기
def calc():
    a = 3
    b = 5
    def mul_add(x):
        return a * x + b
    return mul_add

c = calc() # 3x + 5 -> mul_add
print(c(1), c(2), c(3))
```

```python
# lambda로 클로저 만들기
def calc():
    a = 3
    b = 5
    return lambda x : a * x + b

c = calc()
print(c)
```

```python
# 연습 @@@@@@@@@@@
def counter():
    i = 0
    def count():
        nonlocal i
        i += 1
        return i
    return count
    
    

c = counter()
for i in range(10):
    print(c(), end = ' ')
```

```python
def countdown(n):
    n -= 1
    return n + 1
    

n = int(input())
c = countdown(n)
for i in range(n):
    print(c(), end = ' ')
```
