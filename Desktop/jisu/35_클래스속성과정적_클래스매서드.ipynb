```python
# 인스턴스 속성 
class Person:
    def __init__(self):
        self.bag = []

    def put_bag(self, stuff):
        self.bag.append(stuff)

james = Person()
james.put_bag("책")

maria = Person()
maria.put_bag("열쇠")

print(james.bag)
print(maria.bag)
    
```

```python
## 비공개 클래스 속성 사용하기
# class 클래스이름:
#     __속성 = 값

class Knight:
    __item_limit = 10

    def print_item_limit(self):
        print(Knight.__item_limit)

x = Knight()
print(x.__item_limit)
```

```python
# 정적 매서드 사용하기
# class 클래스이름:
#     @staticmethod
#     def 매서드(매개변수1, 매개변수2):
#         코드
class Calc:
    @staticmethod
    def add(a, b):
        print(a + b)
    
    @staticmethod
    def mul(a, b):
        print(a * b)

print(Calc.add(10, 20))
print(Calc.mul(10, 20)) # return이 없어서 None?
```

```python
## 클래스 매서드 사용하기
# class 클래스이름:
#     @classmethod
#     def 매서드(cls, 매개변수1, 매개변수2):
#         코드
class Person:
    count = 0
    
    def __init__(self):
        Person.count += 3

    @classmethod
    def print_count(cls):
        print(f"{cls.count}명 생성되었습니다")
```

```python
james = Person()
maria = Person()

Person.print_count()
```

```python
## 연습
class Date:
    @ staticmethod
    def is_date_vaild(date):
        year, mon, day = map(int, date.split("-"))
        # print(year, mon, day)
        return 1 <= mon <= 12 and 1 <= day <= 31


if Date.is_date_vaild("2000-10-31"):
    print("올바른 날짜 형식입니다")

else:
    print("잘못된 날짜 형식입니다.")
```

```python
## 심사
class Time:
    def __init__(self, hour, min, sec):
        self.hour = hour
        self.min = min 
        self.sec = sec

        
    @classmethod
    def from_string(cls, time_string):
        hour, min, sec = map(int, time_string.split(":"))
        return cls(hour, min, sec)

    @staticmethod
    def is_time_vaild(time_string):
        hour, min, sec = map(int, time_string.split(":"))
        return 0 <= hour <= 24 and 0 <= min <= 59 and 0 <= sec <= 60

        
            


time_string = input()

if Time.is_time_vaild(time_string):
    t = Time.from_string(time_string)
    print(f"{t.hour} : {t.min} : {t.sec}")

else:
    print("잘못된 시간입니다")
```
