```python
# if, elif, else 모두 사용
# if 조건식:
#     코드1
# elif 조건식:
#     코드2
# else:
#     코드3

x = 30
if x == 10:
    print("10입니다")
    
elif x == 20:
    print("20입니다")

else:
    print("10도 20도 아닙니다")
    
```

```python
# 음료수 자판기 만들기
a = int(input())

if a == 1:
    print("콜라")

elif a == 2:
    print("사이다")

elif a == 3:
    print("환타")

else:
    print("error")
```

```python
num = int(input())
if 11 <= num <= 20:
    print("11~20")
elif 21 <= num <= 30:
    print("21~30") 

else:
    print("아무것도 해당하지 않음")
```

```python
age = int(input("나이 "))
money = int(input("돈 "))

if 7 <= age <= 12:
    print("650원") 
    money -= 650
elif age <= 18:
    print("1,050원")
    money -= 1050
elif 19 >= age:
    print("1,250")
    money -= 1250
else:
    print("error")

print(f"{money}원")
```
