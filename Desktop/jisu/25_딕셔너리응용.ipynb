```python
# 딕셔너리 조작하기
## setdefault, update

a = {
    "a" : 10,
    "b" : 20,
    "c" : 30,
    "d" : 40
}

a.setdefault("e")
print(a)
a.setdefault("f", 100)
print(a)
```

```python
## 두 개의 딕셔너리에서 키의 값 수정
a = {
    "a" : 10,
    "b" : 20,
    "c" : 30,
    "d" : 40
}

a.update(a = 90)
print(a)
```

```python
a.update(e = 90)
print(a)
```

```python
a.update(a = 900, f = 6)
print(a)
```

```python
y = {
    1 : "one",
    2 : "two"
}

y.update(
    {1 : "ONE", 2 : "TWO"}
)

print(y)
```

```python
y.update(
    [
        [2, "two"],
        [4, "four"]
    ]

)

print(y)
```

```python
y.update(
    zip([1, 2], ["one", "two"])
)

print(y)
```

```python
## 딕셔너리 키-값 쌍 삭제하기
x = {
    "a" : 10,
    "b" : 20,
    "c" : 30,
    "d" : 40
}

re = x.pop("a") # pop은 키 값으로만 삭제 가능// + remove
print(re)
x.pop("z", 0) # 없으면 0을 출력하게 함
print(x)
del x["b"] # []로 키를 불러 삭제
print(x)
```

```python
x = {
    "a" : 10,
    "b" : 20,
    "c" : 30,
    "d" : 40
}

re = x.popitem() # 가장 뒤에 있는 것을 삭제, 삭제된 key와 value를 불러올 수 있다
print(x)
print(re)

x.clear()
print(x)
```

```python
## 반복문으로 딕셔너리의 키-값 쌍을 모두 출력하기
x = {
    "a" : 10,
    "b" : 20,
    "c" : 30,
    "d" : 40
}

# for i in x:
#     print(i) # 키의 값만 출력

for key, val in x.items(): # items로 키와 값 모두 출력
    print(f"{key} : {val}")
```

```python
## 딕셔너리 표현식 사용하기

# {키:값 for 키, 값 in 딕셔너리}
# dict({키:값 for 키, 값 in 딕셔너리})

keys = ['a', 'b', 'c', 'd']
imsi = {
    'a' : None,
    'b' : None,
    'c' : None,
    'd' : None

}

# for key, vlaue in dict.formkeys(keys).items()}
x = {key : val for key, val in imsi.items()}
print(x)
```

```python
## 딕셔너리 표현식에서 if문 사용하기
x = {
    "a" : 10,
    "b" : 20,
    "c" : 30,
    "d" : 40
}

# for key, val in x.items(): # 실시간으로 사이즈가 바뀜 사이즈가 달라져서 error// copy를 뜨는게 좋음
for key, val in x.copy().items(): # copy를 해서 실시간으로 변경되지 않음
    if val == 20:
        del x[key]

print(x)
```

```python
x = {
    "a" : 10,
    "b" : 20,
    "c" : 30,
    "d" : 40
}

x = {key : val for key, val in x.items() if val != 20}
print(x)
```

```python
# 심사
a = "alpha bravo charlie delta"
score = "10 20 30 40"

a = list(a.split())
score = list(map(int, score.split()))

dic = dict(zip(
    a, score
))

dic.pop('delta')

for key, val in dic.copy().items():
    if val == 30:
        del dic[key]

print(dic)
```
