```python
# 문자열 판단하기
import re

re.match("hello", "hello, world")
```

```python
print(re.search("hello", "hello, world"))
print(re.search("world!$", "hello, world!"))
```

```python
## 지정된 문자열이 하나라도 포함되는지 판단하기
re.match("hello|world", "hello")
```

```python
# 범위 판단하기
## 사칙연산이 아님
print(re.match("[0-9]*", "1234")) # 없어도 None이 안 나옴
print(re.match("[0-9]*", "abcd")) # 영어라도 None이 안 나옴

print(re.match("[0-9]+", "1234")) # 없으면 None
print(re.match("[0-9]+", "abcd")) # 영어라 None
```

```python
# * >> a부터 b까지인데 a가 없어도 됨
print(re.match('a*b', 'b')) # b만 있어도 가능
print(re.match('a*b', 'aab'))
print(re.match('a*b', 'aaababb'))

# + >> a부터 b까지인데 a가 꼭 있어야함
print(re.match('a+b', 'b')) # b만 있으면 불가능, ab 가능
print(re.match('a+b', 'aaaaaaaaab'))
```

```python
# 문자가 한 개만 있는지 판단하기
# 문자?
# [0-9]?

# . == 하나의 값이 오는 위치
# ? == 

print(re.match("abc?d", "abcd"))
print(re.match("abc?d", "abd"))
print(re.match("ab[0-9]?c", "ab3c"))
print(re.match("ab.d", "abxd"))
```

```python
# 문자 개수 판단하기
# 문자(개수)
# (문자열){개수}

print(re.match('h{3}', "hhhello"))
print(re.match('(hello){3}', "hellohellohelloworld"))
```

```python
# [0-9]{개수}
re.match("[0-9]{3}-[0-9]{4}-[0-9]{4}", '010-1000-1000')
```

```python
## 숫자와 영문 문자를 조합해서 판단하기
# a-z
# A-Z

print(re.match("[a-zA-Z0-9]+", "Hello1234"))
print(re.match("[A-Z0-9]+", "hello1234")) ## 대문자믄 없고, 소문자만 있어서 안됨
```

```python
print(re.match("[가-힣]+", "홍길동")) # 힣이 한글의 마지막 순서
print(re.match("[가-힣]+ [가-힣]+", "대구시 달서구"))
```

```python
# 특정 문자 범위에 포함되지 않는지 판단하기
# [^범위]*
# [^범위]+
print(re.match("[^A-Z]+", "Hello")) # 대문자가 있어서 매칭X
print(re.match("[^A-Z]+", "hello"))
```

```python
# ^[범위]*
# ^[범위]+
print(re.search("^[A-Z]+", "Hello"))
```

```python
# [범위]*$
# [범위]+$
print(re.search("[0-9]*$", "123"))
print(re.search("[0-9]+$", "1234"))
```

```python
# 특수문자 판단하기
# \특수문자
print(re.search("\*+", "1 ** 2"))
print(re.search("[$()a-zA-Z0-9]+", "$(document)"))

```

```python
# \d == [0-9]
# \D == [^0-9]
# \w == [a-zA-Z0-9_]
# \W == [^a-zA-Z0-9_]

print(re.match(r"\d+", "1234"))
print(re.match(r"\D+", "Hello"))
print(re.match(r"\w+", "Hello_1234"))
print(re.match(r"\W+", "(:)"))
```

```python
## 공백 처리하기
# \s, \S
print(re.match("[a-zA-Z0-9 ]+", "123 4")) # 공백까지 처리
print(re.match(r"[a-zA-Z0-9\s]+", "123 4")) # 공백까지 처리
```

```python
# 그룹 사용하기
# (정규표현식)(정규표현식)
m = re.match("([0-9]+) ([0-9]+)", "10 295")
print(m.group(1))
print(type(m.group(1)))
print(m.group(2))
print(m.group())
print(m.group(0))
print(m.groups())
print(type(m.groups()))
```

```python
m = re.match(r"(?P<func>[a-zA-Z_][a-zA-Z0-9_]+)\((?P<arg>\w+)\)", "print(1234)") # func(arg) 
print(m.group("func"))
print(m.group("arg"))
```

```python
## 패턴에 매칭되는 모든 문자열 가져오기

# re.findall("패턴", "문자열")
re.findall('[0-9]+', '1 2 fizz 4 buzz fizz 7 8')
```

```python
# 문자열 바꾸기
# re.sub("패턴", "바꿀문자열", "문자열", "바꿀횟수")

re.sub("apple|orange", "friut", "apple box orange tree")
```

```python
re.sub("[0-9+]", "n", "1 2 Fizz 4 Buzz Fizz 7 8")
```

```python
# 교체함수(매치객체)
# re.sub("패턴", 교체함수, "문자열", 바꿀횟수)

def multiple10(m):
    n = int(m.group())
    return str(n * 10)

re.sub("[0-9]+", multiple10, "1 2 Fizz 4 Buzz Fizz 7 8")
```

```python
re.sub("[0-9]+", lambda m: str(int(m.group()) * 10), "1 2 Fizz 4 Buzz Fizz 7 8")
```

```python
# 연습 

p = re.compile(r"^[a-zA-Z0-9+-_.]+@[a-zA-Z0-9-]+\.[a-zA-Z0-9-.]+$")
# ^[a-zA-Z0-9+-_.] -> 반드시 영어나 숫자 + 문자(+-_.)로 시작
# \. 실제 마침표


email1 = "baejisu0910@naver.com"
email2 = "baejisu0910@naver-com"
# print(p.match(email))
print(p.match(email1) != None, end = "")
print(p.match(email2) != None, end = "")
```
