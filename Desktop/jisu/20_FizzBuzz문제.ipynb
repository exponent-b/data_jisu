```python
# 1부터 100까지 출력
for i in range(1, 101):
    print(i)
    
```

```python
# 3의 배수일 때와 5의 배수일 때

for i in range(1, 101):
    if i % 3 == 0:
        print("fizz")
    
    elif i % 5 == 0:
        print("buzz")

    else:
        print(i)
```

```python
# 3과 5의 공배수 처리하기
for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0:
        print("fizz, buzz")
    
    elif i % 3 == 0:
        print("fizz")
    
    elif i % 5 == 0:
        print("buzz")

    else:
        pass
```

```python
for i in range(1, 101):
    if i % 15 == 0:
        print("fizzbuzz")
    elif i % 5 == 0:
        print("buzz")
    elif i % 3 == 0:
        print("fizz")
```

```python
for i in range(1, 101):
    print("fizz" * (i % 3 == 0) + "buzz" * (i % 5 == 0) or 1)                                                                                               
```

```python
for i in range(1, 101):
    if i % 2 == 0 and i % 11 == 0:
        # or i % 22 == 0:
        print("fizzbuzz")
    elif i % 2 == 0:
        print("fizz")
    elif i % 11 == 0:
        print("buzz")
    else:
        print(i)
```
