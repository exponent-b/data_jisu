```python
# while 반복문 사용하기

# 초기식
# while 조건식:
#     반복할 코드
#     변화식

i = 0
while i < 10:
    print("hello")
    i += 1

for i in range(10):
    print("안녕")
```

```python
# 초깃값을 1부터 시작하기
i = 1
while i <= 10:
    print("hello")
    i += 1
```

```python
i = 10
while i > 0 :
    print("hello world", i)
    i -= 1
```

```python
# 입력한 횟수대로 반복하기
user = int(input("입력할 횟수: "))
i = 0
while i < user:
    print("hello", i)
    i += 1
```

```python
# 반복 횟수가 정해지지 않은 경우
import random

print(int(random.random() * 100))
print(random.randint(1, 6))
```
