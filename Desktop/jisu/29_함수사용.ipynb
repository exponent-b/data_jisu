```python
# 함수에서 값을 여러개 반환하기

def add_sub(a, b):
    return a + b, a - b
```

```python
a, b = add_sub(10, 20)
c = add_sub(20, 20)

print(a, b)
print(type(a))
print(type(b))
print(c)
print(type(c))
```

```python
# 함수의 호출 과정 알아보기
def mul(a, b):
    c = a * b
    return c

def add(a, b):
    c = a + b
    print(c)
    d = mul(a, b) # 함수 안에서 함수 호출
    print(d)

x = 10
y = 20
add(x, y)
```

```python
# 연습
x = 10
y = 3

def get_quotient_remainder(a, b):
    c = a // b
    d = a % b

    return c, d

quotient, remainder = get_quotient_remainder(x, y)
print("몫: {0}, 나머지: {1}".format(quotient, remainder))
```

```python
# 심사
x, y = 10, 20

def calc(x, y):
    a = x + y
    s = x - y
    m = x * y
    d = x / y

    return a, s, m, d

a, s, m, d = calc(x, y)
print("덧셈: {0}, 뺄셈: {1}, 곱셈: {2}, 나눗셈: {3}".format(a, s, m, d))
```
